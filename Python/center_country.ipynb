# 국가별 통역사 필요 현황 조사

## 예시 '베트남'

```python
#베트남
a = df_total[['베트남어', '베트남']]

#열 이름 수정
a = a.rename(columns={'베트남어':'통역사', '베트남':'베트남인'})

#'베트남인' 열 평균값
avg_vit = a['베트남인'].mean() 

#새로운 '필요' 열 초기화
a['필요'] = 0

#2: avg_vit보다 크거나 같고 '통역사' == 1
#1: avg_vit보다 크거나 같은데 '통역사' == 0
#0: avg_vit보다 작은 경우
for i in range(a.shape[0]):
    idx = a.index
    #'베트남인' 값이 평균 이상
    if a.loc[idx[i], '베트남인'] >= avg_vit:
        if a.loc[idx[i], '통역사'] == 0:   #'통역사' 값이 0
            a.loc[idx[i], '필요'] = 1  #'필요' 열에 1 추가
        else:   #'통역사' 값이 1
            a.loc[idx[i], '필요'] = 2  #'필요' 열에 2 추가
    #'베트남인' 값이 평균 미만
    else:
        a.loc[idx[i], '필요'] = 0  #'필요' 열에 0 추가

a.tail()
```

## for문을 활용

```python
#변수로 설정할 국가 네이밍
countries = {'베트남': 'vit', '중국': 'chi', '일본': 'jap', '네팔': 'nel', '몽골': 'mon', '필리핀': 'pil'}

#각 국가에 대해 데이터프레임 생성 및 처리
for i, country in enumerate(countries):
    #i번째 언어 열과 i+6번째 국가 열 선택
    df_country = df_total[[f'{country}어', country]]
    
    #열 이름 수정
    df_country = df_country.rename(columns={f'{country}어': '통역사', country: f'{country}인'})
    
    #국가별 평균값 계산
    avg_country = df_country[f'{country}인'].mean()
    
    #새로운 '필요' 열 초기화
    df_country['필요'] = 0
    
    #'필요' 열 업데이트
    for j in range(df_country.shape[0]):
        idx = df_country.index
        if df_country.loc[idx[j], f'{country}인'] >= avg_country:
            if df_country.loc[idx[j], '통역사'] == 0:
                df_country.loc[idx[j], '필요'] = 1
            else:
                df_country.loc[idx[j], '필요'] = 2
        else:
            df_country.loc[idx[j], '필요'] = 0
    
    #동적 변수 생성: globals() 함수를 사용해 count_{country} 형식으로 변수명을 생성하여 각 국가의 데이터프레임을 저장
    globals()[f'df_{countries[country]}'] = df_country
```

```python
#결과를 저장할 딕셔너리 초기화
count_dfs = {}

#각 국가에 대해 '필요' 값 개수를 세고, 데이터프레임으로 변환
for country in countries:
    #'필요' 열의 값 개수 세기
    count_df = globals()[f'df_{countries[country]}']['필요'].value_counts().reset_index()
    
    #열 이름 변경
    count_df.columns = ['필요', countries[country].capitalize()]
    
    #'필요' 열을 인덱스로 설정
    count_df.set_index('필요', inplace=True)
    
    #딕셔너리에 결과 저장
    count_dfs[countries[country]] = count_df

#여러 나라의 데이터를 하나의 데이터프레임으로 병합
from functools import reduce
final_df = reduce(lambda left, right: left.join(right, how='outer'), count_dfs.values())

#NaN 값을 0으로 채우고 다시 int로 변환
final_df.fillna(0, inplace=True)
final_df = final_df.astype(int)

final_df
```

**'필요'열에 값에 넣을 0, 1, 2의 기준**<br>

1. 국가별 '등록 외국인' 수 ≥ 국가별 '등록 외국인' 수 평균<br>
    ⅰ. `통역사 == 0`' -> `'필요' == 1`<br>
    ⅱ. `통역사 == 1` -> `'필요' == 2`<br>
2. 국가별 '등록 외국인' 수 ＜ 국가별 '등록 외국인' 수 평균<br>
    ⅰ. '통역사' 값 상관없이 `'필요' == 0`

### 주요 분석 내용
**1. 통역사 필요성 집중 지역**<br>
특정 지역(구)에 여러 국가의 외국인들이 많이 거주하고 있으며, 통역사 제공이 절실한 상황이 확인된다. 특히 관악구와 광진구는 다수 국가에서 평균 이상의 외국인들이 거주하고 있지만 통역사가 부족한 상황이다. 관악구는 베트남, 중국, 일본, 네팔, 필리핀 등 5개 국가에서 광진구는 베트남, 일본, 네팔, 몽골 등 4개 국가에서 외국인 수에 비해 통역사가 부족하다.<br>

**2. 특정 국가별 통역사 부족 지역**<br>
일본의 경우 서울 내 여러 구에서 외국인 수가 평균 이상임에도 불구하고, 많은 지역에서 통역사가 부족한 상태이다. 강남구, 마포구, 서대문구, 성북구, 용산구 등 다양한 구에서 통역사가 필요하다. 필리핀 외국인 역시 동대문구, 서대문구, 성북구, 용산구 등 여러 구에서 통역사가 필요하지만 제공되지 않는 상황이다.<br>

**3. 외국인 커뮤니티의 분포**<br>
특정 국가 외국인들은 서울의 특정 지역에 집중적으로 거주하는 경향이 있다. 예를 들어, 일본과 필리핀 국적의 외국인들은 서울의 성북구, 서대문구, 용산구 같은 지역에 많이 거주하는 것으로 보인다. 이들은 언어적 지원이 부족한 상태로 생활해야 하며, 이는 외국인 사회 통합에 큰 영향을 미칠 수 있다.<br>

### 결론 및 제언
**1. 통역사 배치 필요성**<br>
통역사가 없는 상태에서 외국인 거주자 수가 평균 이상인 지역들에 통역사 지원을 강화할 필요가 있다. 특히 관악구, 광진구, 성북구 등 다수 외국인이 집중적으로 거주하는 지역은 다국적 통역사 지원이 시급하다.<br>

**2. 다문화 지원 정책 강화**<br>
특히 일본, 필리핀, 네팔 등 특정 국가의 외국인들이 많이 거주하는 지역에 대해 다문화 지원 프로그램을 강화하고, 지역별로 필요한 언어 지원 서비스를 구체적으로 설계해야 한다.
